```python
import fasttext
import fasttext.util

fasttext.util.download_model('ru', if_exists='ignore')
```

```python
model = fasttext.load_model('cc.ru.300.bin')
```

```python
model.get_nearest_neighbors('маникюр', k=5)
```

```python
def embeddings_ft(data, shape):
    model = fasttext.load_model('cc.ru.300.bin')
    fasttext.util.reduce_model(model, shape)
    
    emb_vector = []
    
    data = data.tolist()
    
    for words in data:
        vec = [model.get_word_vector(word) for word in words.split(' ')][0].tolist()
        emb_vector.append(vec)
    return pd.DataFrame(emb_vector, columns=['ft_{}'.format(feature) for feature in range(1, shape+1)])
```

```python
emb_vec = embeddings_ft(df['custom_position'], 100)
```

```python
emb_vec
```

```python
data = pd.concat([emb_vec, df.drop(columns='custom_position')], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
import xgboost as xgb
import tensorflow_decision_forests as tfdf
from sklearn.metrics import mean_absolute_percentage_error as mape
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns='salary_from'), data['salary_from'])
```

```python
linear_model = LinearRegression().fit(X_train, y_train)
```

```python
preds = linear_model.predict(X_test)
print('MAPE: ', mape(y_test, preds))
```

```python
ridge_model = Ridge().fit(X_train, y_train)
preds = ridge_model.predict(X_test)
print('MAPE: ', mape(y_test, preds))
```

```python
lasso_model = Lasso().fit(X_train, y_train)
preds = lasso_model.predict(X_test)
print('MAPE: ', mape(y_test, preds))
```

```python
rfr_model = RandomForestRegressor(n_jobs=-1).fit(X_train, y_train)
preds = rfr_model.predict(X_test)
print('MAPE: ', mape(y_test, preds))
```

```python
etr_model = ExtraTreesRegressor().fit(X_train, y_train)
preds = etr_model.predict(X_test)
print('MAPE: ', mape(y_test, preds))
```

```python
dtr_model = DecisionTreeRegressor().fit(X_train, y_train)
preds = dtr_model.predict(X_test)
print('MAPE: ', mape(y_test, preds))
```

```python
xgb_model = xgb.XGBRegressor().fit(X_train, y_train)
preds = dtr_model.predict(X_test)
print('MAPE: ', mape(y_test, preds))
```

```python
# Удалим выражения в скобках
```

```python
def del_brackets(data):
    x1 = data.find('(')
    x2 = data.find(')')
    if x1 == -1:
        return data
    elif x2 == -1:
        return data[:x1]
    else:
        return data[:x1]+data[x2:-1]
```

```python
df['custom_position'] = df['custom_position'].apply(del_brackets)
```

```python
def del_slash(data):
    x1 = data.find('/')
    if x1 == -1:
        return data
    else:
        return data[:x1]
```

```python
df['custom_position'] = df['custom_position'].apply(del_slash)
```

```python
emb_vec = embeddings_ft(df['custom_position'], 50)
```

```python
data = pd.concat([emb_vec, df.drop(columns='custom_position')], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns='salary_from'), data['salary_from'], random_state=42)
```

```python
etr_model = ExtraTreesRegressor(random_state=42).fit(X_train, y_train)
preds = etr_model.predict(X_test)
print('MAPE: ', mape(y_test, preds))
```
