Лемматизация

```python
import spacy 
```

```python
nlp = spacy.load("ru_core_news_sm")

def lemm_data(data):
    data = data.split(',')
    lem_text = []
    for el in data:
        x = []
        for token in nlp(el):
            if (
                not token.is_stop and
                not token.is_punct and
                not token.is_digit and
                not token.like_email and
                not token.is_bracket
            ):
                if len(token.lemma_.lower()) != 1:
                    x.append(token.lemma_.lower())
        lem_text.append(' '.join(x))
    return ', '.join(lem_text)
```

```python
df['sum_skills_lemmed'] = df['sum_skills'].apply(lemm_data)
```

```python
df.loc[df[df['sum_skills_lemmed'].isna()].index.tolist(), 'sum_skills'] = 'Отсутствует'
```

```python
df.loc[df[df['sum_skills_lemmed'].isna()].index.tolist(), 'sum_skills_lemmed'] = 'отсутствовать'
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
import xgboost as xgb
import tensorflow_decision_forests as tfdf
from sklearn.metrics import mean_absolute_percentage_error as mape
from catboost import CatBoostRegressor
from catboost import Pool
```

```python
df_preprocessed = pd.concat([df.drop(columns=['custom_position', 'schedule', 'education_name', 'city_id', 'required_experience']), 
          pd.get_dummies(df[['schedule', 'education_name', 'city_id', 'required_experience']], drop_first=True)],
          axis = 1
         )
```

```python
df_preprocessed
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_preprocessed.drop(columns=['salary_from', 'sum_skills', 'sum_skills_lemmed']),
                                                    df_preprocessed['salary_from'], train_size=0.8, random_state=42)
```

```python
train_pool = Pool(data=X_train, label=y_train, text_features=['skill_'+str(i+1) for i in range(10)])
test_pool = Pool(data=X_test, label=y_test, text_features=['skill_'+str(i+1) for i in range(10)])
```

```python
catboost_model_1 = CatBoostRegressor(
    iterations=1000,
    thread_count=-1
)
```

```python
catboost_model_1.fit(train_pool, eval_set=test_pool)
```

```python
preds = catboost_model_1.predict(test_pool)
```

```python
mape(y_test, preds)
```

```python
pd.DataFrame(X_test.columns, catboost_model_1.get_feature_importance())
```
